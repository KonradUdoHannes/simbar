```python
import pandas as pd
import mplfinance as mpf
import numpy as np
from datetime import datetime,timedelta
```

```python
xs = np.random.randn(1000)
# idx = pd.date_range(start = datetime(2018,1,1),periods = 1000)
df = pd.DataFrame(xs,columns = ['ticks'])#,index=idx)
```

```python
plen = 100
ohlc = df.reset_index().assign(index = lambda df:df.index.map(lambda x : x//plen)).groupby('index').aggregate(Low = ('ticks','min'),High = ('ticks','max'),Open=('ticks','first'),Close=('ticks','last')).set_index(pd.date_range(start = datetime(2018,1,1),periods = 1000//plen))
# ohlc.set_index(pd.date_range(start = datetime(2018,1,1),periods = 250))
```

```python
mpf.plot(ohlc)
```
